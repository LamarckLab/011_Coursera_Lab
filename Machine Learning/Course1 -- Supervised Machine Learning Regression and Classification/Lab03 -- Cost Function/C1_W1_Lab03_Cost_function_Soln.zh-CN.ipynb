## 计算代价（Cost）
本作业中 'cost'（代价）这个词可能有点绕，因为数据本身就是房价（housing cost）。这里的 cost 衡量的是模型对房屋目标价格的预测有多好；而 'price' 专指房价数据。

单变量的代价方程为：
  $$J(w,b) = \frac{1}{2m} \sum\limits_{i = 0}^{m-1} (f_{w,b}(x^{(i)}) - y^{(i)})^2 \tag{1}$$

其中
  $$f_{w,b}(x^{(i)}) = wx^{(i)} + b \tag{2}$$

- $f_{w,b}(x^{(i)})$ 是用参数 $w,b$ 对样本 $i$ 的预测值。
- $(f_{w,b}(x^{(i)}) -y^{(i)})^2$ 是目标值与预测值之差的平方。
- 把所有 $m$ 个样本的这些差求和、再除以 `2m`，就得到代价 $J(w,b)$。
>注意：课程中求和范围通常写作 1 到 m，而代码里是 0 到 m-1。

下面的代码通过遍历每个样本来计算代价。每次循环中：
- 计算预测值 `f_wb`
- 计算目标值与预测值之差并平方
- 把它累加到总代价上

```python
def compute_cost(x, y, w, b): 
    """
    Computes the cost function for linear regression.
    
    Args:
      x (ndarray (m,)): Data, m examples 
      y (ndarray (m,)): target values
      w,b (scalar)    : model parameters  
    
    Returns
        total_cost (float): The cost of using w,b as the parameters for linear regression
               to fit the data points in x and y
    """
    # number of training examples
    m = x.shape[0] 
    
    cost_sum = 0 
    for i in range(m): 
        f_wb = w * x[i] + b   
        cost = (f_wb - y[i]) ** 2  
        cost_sum = cost_sum + cost  
    total_cost = (1 / (2 * m)) * cost_sum  

    return total_cost
```

## 代价函数的直观理解

<img align="left" src="./images/C1_W1_Lab02_GoalOfRegression.PNG"    style=" width:380px; padding: 10px;  " /> 你的目标是找到一个模型 $f_{w,b}(x) = wx + b$，其参数为 $w,b$，能在给定输入 $x$ 时准确预测房价。代价衡量模型在训练数据上的准确程度。

上面的代价方程 (1) 表明：如果能选出合适的 $w$、$b$，使预测 $f_{w,b}(x)$ 与目标 $y$ 相符，那么 $(f_{w,b}(x^{(i)}) - y^{(i)})^2$ 项就为零、代价最小。在这个简单的两点例子里，你可以做到这一点！

上一个实验里你已确定 $b=100$ 是最优解，所以这里把 $b$ 固定为 100，只关注 $w$。

- 代价在 $w = 200$ 时最小，与上一个实验的结果一致
- 由于代价方程里对"目标与预测之差"取了平方，当 $w$ 过大或过小时，代价都会迅速上升
- 用最小化代价选出的 `w` 和 `b`，得到的直线能完美拟合数据
