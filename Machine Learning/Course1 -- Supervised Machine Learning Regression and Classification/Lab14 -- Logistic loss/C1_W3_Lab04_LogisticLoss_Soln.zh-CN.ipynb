## 逻辑回归能用平方误差吗？
<img align="left" src="./images/C1_W3_SqErrorVsLogistic.png"     style=" width:400px; padding: 10px; " > 回忆一下，对 **线性** 回归我们用的是 **平方误差代价函数**：
单变量平方误差代价的方程为：
  $$J(w,b) = \frac{1}{2m} \sum\limits_{i = 0}^{m-1} (f_{w,b}(x^{(i)}) - y^{(i)})^2 \tag{1}$$

其中
  $$f_{w,b}(x^{(i)}) = wx^{(i)} + b \tag{2}$$

回忆一下，平方误差代价有个很好的性质：沿着代价的导数走就能到达最小值。

```python
soup_bowl()
```

这个代价函数在线性回归上效果很好，很自然会想到把它也用于逻辑回归。然而，正如上面幻灯片所指出的，$f_{wb}(x)$ 现在带了一个非线性成分——sigmoid 函数：$f_{w,b}(x^{(i)}) = sigmoid(wx^{(i)} + b )$。我们把平方误差代价用到前面某个实验的例子上（这次带上 sigmoid）试试。

这是我们的训练数据：

```python
x_train = np.array([0., 1, 2, 3, 4, 5],dtype=np.longdouble)
y_train = np.array([0,  0, 0, 1, 1, 1],dtype=np.longdouble)
plt_simple_example(x_train, y_train)
```

现在，我们用 *平方误差代价* 画出代价的曲面图：
  $$J(w,b) = \frac{1}{2m} \sum\limits_{i = 0}^{m-1} (f_{w,b}(x^{(i)}) - y^{(i)})^2 $$

其中
  $$f_{w,b}(x^{(i)}) = sigmoid(wx^{(i)} + b )$$

```python
plt.close('all')
plt_logistic_squared_error(x_train,y_train)
plt.show()
```

虽然这产生了一张挺有意思的图，但上面的曲面远不如线性回归那个 '汤碗' 光滑！

逻辑回归需要一个更适合其非线性本质的代价函数。这要从损失（Loss）函数说起，下面描述。

## 逻辑损失函数（Logistic Loss Function）
<img align="left" src="./images/C1_W3_LogisticLoss_a.png"     style=" width:250px; padding: 2px; " >
<img align="left" src="./images/C1_W3_LogisticLoss_b.png"     style=" width:250px; padding: 2px; " >
<img align="left" src="./images/C1_W3_LogisticLoss_c.png"     style=" width:250px; padding: 2px; " > 

逻辑回归用一个更适合分类任务的损失函数——分类任务的目标是 0 或 1，而非任意数值。

>**定义说明：** 本课程采用如下定义：
**Loss（损失）** 衡量单个样本与其目标值之差，而
**Cost（代价）** 衡量整个训练集上的损失

定义如下：
* $loss(f_{\mathbf{w},b}(\mathbf{x}^{(i)}), y^{(i)})$ 是单个数据点的代价，即：

\begin{equation}
  loss(f_{\mathbf{w},b}(\mathbf{x}^{(i)}), y^{(i)}) = \begin{cases}
    - \log\left(f_{\mathbf{w},b}\left( \mathbf{x}^{(i)} \right) \right) & \text{if $y^{(i)}=1$}\\
    - \log \left( 1 - f_{\mathbf{w},b}\left( \mathbf{x}^{(i)} \right) \right) & \text{if $y^{(i)}=0$}
  \end{cases}
\end{equation}

*  $f_{\mathbf{w},b}(\mathbf{x}^{(i)})$ 是模型的预测值，$y^{(i)}$ 是目标值。

*  $f_{\mathbf{w},b}(\mathbf{x}^{(i)}) = g(\mathbf{w} \cdot\mathbf{x}^{(i)}+b)$，其中函数 $g$ 是 sigmoid 函数。

这个损失函数的关键特征是：它用了两条不同的曲线，一条对应目标为零（$y=0$），另一条对应目标为一（$y=1$）。两者结合，提供了损失函数所需要的行为——预测与目标相符时为零，预测与目标相差越大值上升越快。看看下面的曲线：

```python
plt_two_logistic_loss_curves()
```

合在一起，这些曲线与平方误差损失的二次曲线相似。注意，x 轴是 $f_{\mathbf{w},b}$，即 sigmoid 的输出，严格落在 0 到 1 之间。

上面的损失函数可以改写成更易实现的形式。
    $$loss(f_{\mathbf{w},b}(\mathbf{x}^{(i)}), y^{(i)}) = (-y^{(i)} \log\left(f_{\mathbf{w},b}\left( \mathbf{x}^{(i)} \right) \right) - \left( 1 - y^{(i)}\right) \log \left( 1 - f_{\mathbf{w},b}\left( \mathbf{x}^{(i)} \right) \right)$$

这个方程看起来挺吓人。但一想到 $y^{(i)}$ 只能取 0 和 1 两个值，就没那么可怕了。可以把方程分两种情况看：
当 $ y^{(i)} = 0$ 时，左边一项被消去：
$$
\begin{align}
loss(f_{\mathbf{w},b}(\mathbf{x}^{(i)}), 0) &= (-(0) \log\left(f_{\mathbf{w},b}\left( \mathbf{x}^{(i)} \right) \right) - \left( 1 - 0\right) \log \left( 1 - f_{\mathbf{w},b}\left( \mathbf{x}^{(i)} \right) \right) \\
&= -\log \left( 1 - f_{\mathbf{w},b}\left( \mathbf{x}^{(i)} \right) \right)
\end{align}
$$
当 $ y^{(i)} = 1$ 时，右边一项被消去：
$$
\begin{align}
  loss(f_{\mathbf{w},b}(\mathbf{x}^{(i)}), 1) &=  (-(1) \log\left(f_{\mathbf{w},b}\left( \mathbf{x}^{(i)} \right) \right) - \left( 1 - 1\right) \log \left( 1 - f_{\mathbf{w},b}\left( \mathbf{x}^{(i)} \right) \right)\\
  &=  -\log\left(f_{\mathbf{w},b}\left( \mathbf{x}^{(i)} \right) \right)
\end{align}
$$

好，有了这个新的逻辑损失函数，就能构造出一个把所有样本的损失都纳入的代价函数——这是下一个实验的主题。现在，我们先看看上面那个简单例子的"代价 vs 参数"曲线：

```python
plt.close('all')
cst = plt_logistic_cost(x_train,y_train)
```

这条曲线非常适合梯度下降！它没有平台、没有局部极小值、也没有间断。注意，它不像平方误差那样是碗形。图里同时画了代价和代价的对数，以说明：当代价很小时曲线仍有斜率、会继续下降。提示：你可以用鼠标旋转上面的图。
