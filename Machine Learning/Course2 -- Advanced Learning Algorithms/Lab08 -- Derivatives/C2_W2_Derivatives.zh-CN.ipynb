## 导数的非正式定义

导数的正式定义带着极限和 '趋于零' 的值，可能有点吓人。但其思想其实简单得多。

一个函数的导数，描述的是：当输入变量有一个小变化时，函数的输出如何变化。

我们以代价函数 $J(w)$ 为例。代价 $J$ 是输出，$w$ 是输入变量。
我们给 '小变化' 起个名字 *epsilon* 或 $\epsilon$。用这些希腊字母是因为数学里传统上用 *epsilon*($\epsilon$) 或 *delta* ($\Delta$) 表示一个小值。你可以把它想成 0.001 或别的什么小值。

$$
\begin{equation}
\text{if } w \uparrow \epsilon \text{ causes }J(w) \uparrow \text{by }k \times \epsilon \text{ then}  \\
\frac{\partial J(w)}{\partial w} = k \tag{1}
\end{equation}
$$

这只是说：如果你把函数 $J(w)$ 的输入改变一点点，而输出变化了那一点点的 $k$ 倍，那么 $J(w)$ 的导数就等于 $k$。

我们来试试。看看函数 $J(w) = w^2$ 在 $w=3$、$\epsilon = 0.001$ 处的导数

````python
J = (3)**2
J_epsilon = (3 + 0.001)**2
k = (J_epsilon - J)/0.001    # difference divided by epsilon
print(f"J = {J}, J_epsilon = {J_epsilon}, dJ_dw ~= k = {k:0.6f} ")
````

我们把输入值增加了一点点（0.001），使输出从 9 变成 9.006001，增量是输入增量的 6 倍。参照上面 (1)，这说明 $k=6$，所以 $\frac{\partial J(w)}{\partial w} \approx 6$。如果你熟悉微积分，就知道用符号写作 $\frac{\partial J(w)}{\partial w} = 2 w$，在 $w=3$ 时即为 6。我们上面的计算不恰好是 6，因为要完全正确，$\epsilon$ 需要[无穷小](https://www.dictionary.com/browse/infinitesimally)、或者说非常非常小。这就是我们用 $\approx$ 或 ~= 而非 = 的原因。我们看看把 $\epsilon$ 变小会怎样。

````python
J = (3)**2
J_epsilon = (3 + 0.000000001)**2
k = (J_epsilon - J)/0.000000001
print(f"J = {J}, J_epsilon = {J_epsilon}, dJ_dw ~= k = {k} ")
````

随着我们减小 $\epsilon$ 的大小，这个值接近恰好为 6。你可以试着进一步减小它。

## 求符号导数
在反向传播中，知道简单函数在任意输入值处的导数很有用。换句话说，我们想知道 '符号（symbolic）' 导数、而非 '算术（arithmetic）' 导数。符号导数的一个例子是 $\frac{\partial J(w)}{\partial w} = 2 w$，即上面 $J(w) = w^2$ 的导数。有了符号导数，你就能求出任意输入值 $w$ 处导数的值。

如果你上过微积分课，就熟悉数学家发展出的许多[求导法则](https://en.wikipedia.org/wiki/Differentiation_rules#Power_laws,_polynomials,_quotients,_and_reciprocals)，用来对一个表达式求导。事实上，这个过程已被符号微分程序自动化了。Python 里的一个例子是 [SymPy](https://www.sympy.org/en/index.html) 库。我们看看怎么用它。

### $J = w^2$
定义 python 变量及其符号名。

````python
J, w = symbols('J, w')
````

定义并打印表达式。注意 SymPy 会生成一个 [latex](https://en.wikibooks.org/wiki/LaTeX/Mathematics) 字符串，从而渲染出漂亮易读的方程。

````python
J=w**2
J
````

用 SymPy 的 `diff` 对 $J$ 关于 $w$ 求导。注意结果与我们前面的例子一致。

````python
dJ_dw = diff(J,w)
dJ_dw
````

通过把符号值 '替换（substitute）' 成数值，在几个点上求导数的值。第一个例子里，$w$ 被替换成 $2$。

````python
dJ_dw.subs([(w,2)])    # derivative at the point w = 2
````

````python
dJ_dw.subs([(w,3)])    # derivative at the point w = 3
````

````python
dJ_dw.subs([(w,-3)])    # derivative at the point w = -3
````

## $J = 2w$

````python
w, J = symbols('w, J')
````

````python
J = 2 * w
J
````

````python
dJ_dw = diff(J,w)
dJ_dw
````

````python
dJ_dw.subs([(w,-3)])    # derivative at the point w = -3
````

把它与算术计算作比较

````python
J = 2*3
J_epsilon = 2*(3 + 0.001)
k = (J_epsilon - J)/0.001
print(f"J = {J}, J_epsilon = {J_epsilon}, dJ_dw ~= k = {k} ")
````

对函数 $J=2w$，很容易看出：无论 $w$ 的起始值是多少，$w$ 的任何变化都会使输出 $J$ 变化其 2 倍。我们的 NumPy 和算术结果证实了这一点。

## $J = w^3$

````python
J, w = symbols('J, w')
````

````python
J=w**3
J
````

````python
dJ_dw = diff(J,w)
dJ_dw
````

````python
dJ_dw.subs([(w,2)])   # derivative at the point w=2
````

把它与算术计算作比较

````python
J = (2)**3
J_epsilon = (2+0.001)**3
k = (J_epsilon - J)/0.001
print(f"J = {J}, J_epsilon = {J_epsilon}, dJ_dw ~= k = {k} ")
````

## $J = \frac{1}{w}$

````python
J, w = symbols('J, w')
````

````python
J= 1/w
J
````

````python
dJ_dw = diff(J,w)
dJ_dw
````

````python
dJ_dw.subs([(w,2)])
````

把它与算术计算作比较

````python
J = 1/2
J_epsilon = 1/(2+0.001)
k = (J_epsilon - J)/0.001
print(f"J = {J}, J_epsilon = {J_epsilon}, dJ_dw ~= k = {k} ")
````
