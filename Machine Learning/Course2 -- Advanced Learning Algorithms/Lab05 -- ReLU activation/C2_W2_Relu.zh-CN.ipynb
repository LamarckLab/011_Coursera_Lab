右边课程里的例子展示了 ReLU 的一个应用。此例中，导出的 "awareness" 特征不是二值的，而是有连续的取值范围。sigmoid 最适合 on/off 或二值情形，而 ReLU 提供一种连续的线性关系；此外它还有一段输出为零的 'off' 区间。
这个 "off" 特性使 ReLU 成为一个非线性激活。为什么需要它？我们在下面考察一下。

### 为什么需要非线性激活？
<img align="left" src="./images/C2_W2_ReLU_Graph.png"     style=" width:250px; padding: 10px 20px; " > 图中所示的函数由若干线性片段组成（分段线性 piecewise linear）。在线性部分内斜率保持一致，在转折点处突然改变。在转折点，会加入一个新的线性函数，它与已有函数相加后产生新的斜率；这个新函数在转折点处加入，但在该点之前对输出没有贡献。非线性激活函数负责在转折点之前（有时之后）关闭输入。下面的练习给出一个更具体的例子。


这个练习的目的是体会 ReLU 的非线性行为如何提供"在需要之前关闭函数"这一必要能力。我们看看它在这个例子里是怎么起作用的。
<img align="right" src="./images/C2_W2_ReLU_Plot.png"     style=" width:600px; padding: 10px 20px; ">
右边的图是第一层各 unit 的输出。
从上往下看，unit 0 负责标记为 1 的第一段。图中同时画了线性函数 $z$ 和经过 ReLU 后的函数 $a$。可以看到 ReLU 在区间 [0,1] 之后把函数截断，这很重要，因为它防止 unit 0 干扰后面的段。

unit 1 负责第 2 段。这里 ReLU 让这个 unit 在 x 超过 1 之前保持沉默。由于第一个 unit 没有贡献，unit 1 的斜率 $w^{[1]}_1$ 就是目标线的斜率。偏置必须调整，使输出在 x 到达 1 之前保持为负。注意 unit 1 的贡献也延伸到第 3 段。

unit 2 负责第 3 段。ReLU 再次把输出置零，直到 x 达到合适的值。这个 unit 的斜率 $w^{[1]}_2$ 必须设定，使 unit 1 和 unit 2 之和具有所需的斜率。偏置同样被调整，使输出在 x 到达 2 之前保持为负。

ReLU 激活的 "off"（关闭）特性，让模型能把一段段线性片段拼接起来，从而建模复杂的非线性函数。
