# 线性回归的闭式解（closed-form solution）
Scikit-learn 有一个[线性回归模型](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html#sklearn.linear_model.LinearRegression)，实现了闭式（closed-form）线性回归。

我们用前面实验的数据——一套 1000 sqft 的房子卖了 \$300,000，一套 2000 sqft 的房子卖了 \$500,000。

| Size (1000 sqft)     | Price (1000s of dollars) |
| ----------------| ------------------------ |
| 1               | 300                      |
| 2               | 500                      |

### 加载数据集

```python
X_train = np.array([1.0, 2.0])   #features
y_train = np.array([300, 500])   #target value
```

### 创建并拟合模型
下面的代码用 scikit-learn 做回归。
第一步创建一个回归对象。
第二步调用该对象的一个方法 `fit`，它执行回归、把参数拟合到输入数据上。该工具包期望一个二维的 X 矩阵。

```python
linear_model = LinearRegression()
#X must be a 2-D Matrix
linear_model.fit(X_train.reshape(-1, 1), y_train) 
```

### 查看参数
在 scikit-learn 中，$\mathbf{w}$ 和 $\mathbf{b}$ 参数分别称为 'coefficients'（系数）和 'intercept'（截距）。

```python
b = linear_model.intercept_
w = linear_model.coef_
print(f"w = {w:}, b = {b:0.2f}")
print(f"'manual' prediction: f_wb = wx+b : {1200*w + b}")
```

### 做预测

调用 `predict` 函数即可生成预测。

```python
y_pred = linear_model.predict(X_train.reshape(-1, 1))

print("Prediction on training set:", y_pred)

X_test = np.array([[1200]])
print(f"Prediction for 1200 sqft house: ${linear_model.predict(X_test)[0]:0.2f}")
```

## 第二个例子
第二个例子来自前面一个多特征的实验。最终的参数值和预测，与那个实验里未归一化的"长运行"结果非常接近。那次未归一化的运行要花几个小时出结果，而这里几乎是瞬间完成。闭式解在这种较小的数据集上效果很好，但在更大的数据集上计算开销可能很大。
>闭式解不需要归一化。

```python
# load the dataset
X_train, y_train = load_house_data()
X_features = ['size(sqft)','bedrooms','floors','age']
```

```python
linear_model = LinearRegression()
linear_model.fit(X_train, y_train) 
```

```python
b = linear_model.intercept_
w = linear_model.coef_
print(f"w = {w:}, b = {b:0.2f}")
```

```python
print(f"Prediction on training set:\n {linear_model.predict(X_train)[:4]}" )
print(f"prediction using w,b:\n {(X_train @ w + b)[:4]}")
print(f"Target values \n {y_train[:4]}")

x_house = np.array([1200, 3,1, 40]).reshape(-1,4)
x_house_predict = linear_model.predict(x_house)[0]
print(f" predicted price of a house with 1200 sqft, 3 bedrooms, 1 floor, 40 years old = ${x_house_predict*1000:0.2f}")
```
