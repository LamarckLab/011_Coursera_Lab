## 数据集

假设你有以下训练数据集
- 输入变量 `X` 是一个 numpy 数组，有 6 个训练样本，每个样本两个特征
- 输出变量 `y` 也是一个 numpy 数组，有 6 个样本，`y` 取 `0` 或 `1`

```python
X = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y = np.array([0, 0, 0, 1, 1, 1]).reshape(-1,1) 
```

### 画出数据

我们用一个辅助函数把数据画出来。标签 $y=1$ 的数据点显示为红色叉号，标签 $y=0$ 的数据点显示为蓝色圆圈。

```python
fig,ax = plt.subplots(1,1,figsize=(4,4))
plot_data(X, y, ax)

ax.axis([0, 4, 0, 3.5])
ax.set_ylabel('$x_1$')
ax.set_xlabel('$x_0$')
plt.show()
```

## 逻辑回归模型


* 假设你想在这份数据上训练一个如下形式的逻辑回归模型

  $f(x) = g(w_0x_0+w_1x_1 + b)$

  其中 $g(z) = \frac{1}{1+e^{-z}}$，即 sigmoid 函数


* 假设你训练了模型，得到参数 $b = -3, w_0 = 1, w_1 = 1$。也就是

  $f(x) = g(x_0+x_1-3)$

  （如何把这些参数拟合到数据，你会在课程后面学到）


我们通过画出决策边界，来理解这个训练好的模型在预测什么

### 逻辑回归与决策边界回顾

* 回忆一下，逻辑回归的模型表示为

  $$f_{\mathbf{w},b}(\mathbf{x}^{(i)}) = g(\mathbf{w} \cdot \mathbf{x}^{(i)} + b) \tag{1}$$

  其中 $g(z)$ 是 sigmoid 函数，它把所有输入值映射到 0 到 1 之间：

  $g(z) = \frac{1}{1+e^{-z}}\tag{2}$
  而 $\mathbf{w} \cdot \mathbf{x}$ 是向量点积：

  $$\mathbf{w} \cdot \mathbf{x} = w_0 x_0 + w_1 x_1$$


 * 我们把模型的输出 ($f_{\mathbf{w},b}(x)$) 解释为：在给定 $\mathbf{x}$、由 $\mathbf{w}$ 和 $b$ 参数化下，$y=1$ 的概率。
* 因此，要从逻辑回归模型得到最终预测（$y=0$ 或 $y=1$），可以用如下判据——

  若 $f_{\mathbf{w},b}(x) >= 0.5$，预测 $y=1$

  若 $f_{\mathbf{w},b}(x) < 0.5$，预测 $y=0$


* 我们把 sigmoid 函数画出来，看看 $g(z) >= 0.5$ 在哪里

```python
# Plot sigmoid(z) over a range of values from -10 to 10
z = np.arange(-10,11)

fig,ax = plt.subplots(1,1,figsize=(5,3))
# Plot z vs sigmoid(z)
ax.plot(z, sigmoid(z), c="b")

ax.set_title("Sigmoid function")
ax.set_ylabel('sigmoid(z)')
ax.set_xlabel('z')
draw_vthresh(ax,0)
```

* 可以看到，当 $z >=0$ 时 $g(z) >= 0.5$

* 对逻辑回归模型，$z = \mathbf{w} \cdot \mathbf{x} + b$。因此，

  若 $\mathbf{w} \cdot \mathbf{x} + b >= 0$，模型预测 $y=1$

  若 $\mathbf{w} \cdot \mathbf{x} + b < 0$，模型预测 $y=0$



### 画出决策边界

现在回到我们的例子，理解逻辑回归模型是怎么做预测的。

* 我们的逻辑回归模型形如

  $f(\mathbf{x}) = g(-3 + x_0+x_1)$


* 由上面所学，可以看出：当 $-3 + x_0+x_1 >= 0$ 时，该模型预测 $y=1$

我们从图形上看看它是什么样。先画出 $-3 + x_0+x_1 = 0$，它等价于 $x_1 = 3 - x_0$。

```python
# Choose values between 0 and 6
x0 = np.arange(0,6)

x1 = 3 - x0
fig,ax = plt.subplots(1,1,figsize=(5,4))
# Plot the decision boundary
ax.plot(x0,x1, c="b")
ax.axis([0, 4, 0, 3.5])

# Fill the region below the line
ax.fill_between(x0,x1, alpha=0.2)

# Plot the original data
plot_data(X,y,ax)
ax.set_ylabel(r'$x_1$')
ax.set_xlabel(r'$x_0$')
plt.show()
```

* 上图中，蓝线代表直线 $x_0 + x_1 - 3 = 0$，它应在 x1 轴上交于 3（若设 $x_1$ = 3、$x_0$ = 0），在 x0 轴上交于 3（若设 $x_1$ = 0、$x_0$ = 3）。

* 阴影区域代表 $-3 + x_0+x_1 < 0$，直线上方的区域是 $-3 + x_0+x_1 > 0$。

* 阴影区域（直线下方）内的任意点被分类为 $y=0$；直线上或直线上方的任意点被分类为 $y=1$。这条线就叫 "决策边界（decision boundary）"。

正如课程所见，通过使用更高阶的多项式项（如 $f(x) = g( x_0^2 + x_1 -1)$），我们能得到更复杂的非线性边界。
