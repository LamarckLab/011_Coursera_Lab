# Ungraded Lab: 用 Scikit-Learn 做逻辑回归

## 目标
在本实验中，你将：
- 用 scikit-learn 训练一个逻辑回归模型。

## 数据集
我们从和之前一样的数据集开始。

```python
import numpy as np

X = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y = np.array([0, 0, 0, 1, 1, 1])
```

## 拟合模型

下面的代码从 scikit-learn 导入[逻辑回归模型](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html#sklearn.linear_model.LogisticRegression)。调用 `fit` 函数即可在训练数据上拟合这个模型。

```python
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression()
lr_model.fit(X, y)
```

## 做预测

调用 `predict` 函数即可查看这个模型做出的预测。

```python
y_pred = lr_model.predict(X)

print("Prediction on training set:", y_pred)
```

## 计算准确率

调用 `score` 函数即可计算这个模型的准确率。

```python
print("Accuracy on training set:", lr_model.score(X, y))
```
